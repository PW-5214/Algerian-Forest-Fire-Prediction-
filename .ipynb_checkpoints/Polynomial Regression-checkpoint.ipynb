```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import sklearn as sk
```

```python
x = 6*np.random.rand(100,1)- 3
y = 0.5*x**2 + 1.5*x + 2 + np.random.randn(100,1)
## eqn : y=0.5x^2 +1.5x+ 2 + outliers
```

```python
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8,random_state=42)
```

##### Simple Linear Reg for Checck min R2 score 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2  = r2_score(y_test,y_pred)
```

```python
r2
```

```python
plt.plot(x_train,model.predict(x_train))
plt.scatter(x_train,y_train)
```

### Pipeline Concepts

```python
from sklearn.pipeline import Pipeline
```

```python

def poly_reg(degree):
    x_new = np.linspace(-3, 3, 200).reshape(200, 1)
    poly_features = PolynomialFeatures(degree=degree, include_bias=True)
    lin_reg = LinearRegression()
    poly_reg = Pipeline([
        ("poly_features", poly_features),
        ("lin_reg", lin_reg)
    ])              ### 1st poly_featuring exec then linear_reg exec step by step execution  it is the pipeline
    poly_reg.fit(x_train, y_train)
    y_pred_new = poly_reg.predict(x_new)
    plt.plot(x_new, y_pred_new, "r")      # Fixed here
    plt.plot(x_train, y_train, "b.")
    plt.plot(x_test, y_test, "g.")
    plt.title(f"Polynomial Regression (degree={degree})")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
```

```python
poly_reg(2)
```
