```python
### independent and dependemt 

x  = df.drop('FWI',axis =1)
y = df['FWI']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
### feature select based on the Coreelation
df.corr()
```

```python
## check multicolinerarity
plt.figure(figsize=(12,7))
corr = x_train.corr()
sns.heatmap(corr,annot = True)
```

```python
def correlation(dataset,thrs):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > thrs:         # iloc[ like FWI and ISI] > 0.80
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
                
```

```python
## threshold set
cf = correlation(x_train,0.85)
```

```python
## drop this ehen corr is greater than 0.85
x_train.drop(cf,axis=1,inplace=True)
x_test.drop(cf,axis=1,inplace=True)
x_train.shape,x_test.shape
```

#### Feature Scaling and Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x_test_sc
```

```python
## check outlers to check effect of the standardization

plt.subplots(figsize = (15,5))
plt.subplot(1,2,1)
sns.boxplot(data= x_train)
plt.title("X_train before scaling")
plt.subplot(1,2,2)
sns.boxplot(data= x_train_sc)
plt.title("X_train after scaling")
```
